### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "resources/city_data.csv"
ride_data_to_load = "resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)

#import dataframe to image 
import dataframe_image as dfi 
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
# pyber_data_df[pyber_data_df['city'] == "Lake Jonathanshire"]

```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
# groupby on the "type" column and count the ride_id 
total_rides_city_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides_city_type
```

```python
# 2. Get the total drivers for each city type
# also use groupby on the "type" column and count() method on the "driver_count"

# driver_per_city_type = pyber_data_df.groupby(["type"]).sum()["driver_count"]
# driver_per_city_type

driver_per_city_type = city_data_df.groupby('type').sum()['driver_count']
driver_per_city_type
```

```python
#  3. Get the total amount of fares for each city type
# use groupby on the type column and sum() on the fares 
tot_fares_per_city = pyber_data_df.groupby(["type"]).sum()["fare"]
tot_fares_per_city

# cant use count() bc we are trying to add the values of each column
```

```python
#  4. Get the average fare per ride for each city type. 
# use mean() 
avg_fare_city_type = pyber_data_df.groupby(["type"]).mean()["fare"]
avg_fare_city_type
```

```python
# 5. Get the average fare per driver for each city type. 
# groupby city 'type' then sum() the cost for each city type 
# divide that by the drivers per city type, 'driver_per_city_type'

avg_fare_per_driver_city_type = pyber_data_df.groupby(["type"]).sum()["fare"] / driver_per_city_type
avg_fare_per_driver_city_type
```

```python
#  6. Create a PyBer summary DataFrame. 
summary_df = pd.DataFrame({'total rides':total_rides_city_type, 
                          'total drivers':driver_per_city_type,
                          'total fares':tot_fares_per_city,
                           'average fare per ride':avg_fare_city_type,
                           'average fare per driver':avg_fare_per_driver_city_type})
summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
summary_df.index.name = None
summary_df
```

```python
#  8. Format the columns.
# use the map({}) and format function 
summary_df['total rides'] = summary_df['total rides'].map('{:,}'.format)
summary_df['total drivers'] = summary_df['total drivers'].map('{:,}'.format)
summary_df['total fares'] = summary_df['total fares'].map('${:,.2f}'.format)
summary_df['average fare per ride'] = summary_df['average fare per ride'].map('${:,.2f}'.format)
summary_df['average fare per driver'] = summary_df['average fare per driver'].map('${:,.2f}'.format)

summary_df
```

```python
dfi.export(summary_df, 'analysis/summary_df.png')
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
# groupby parameters are city type and date 
fares_date_city_type_df = pyber_data_df.groupby(["type","date"]).sum()["fare"]
fares_date_city_type_df
                                                 
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
fares_date_city_type_df = fares_date_city_type_df.reset_index()
fares_date_city_type_df.head()
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
# use the pivot() method to create the table 
fares_date_city_type_df = fares_date_city_type_df.pivot(index = 'date', columns = 'type', values = 'fare')
fares_date_city_type_df.head()
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
# loc() access a group of rows & column by labels 

dates_fares_df = fares_date_city_type_df.loc['2019-01-01':'2019-04-29']
dates_fares_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)

dates_fares_df.index = pd.to_datetime(dates_fares_df.index)
dates_fares_df.head()
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
dates_fares_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
dates_fares_df = dates_fares_df.resample('W').sum()
dates_fares_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

dates_fares_df.plot(figsize = (14,5),)
plt.xlabel('')
plt.ylabel('Fare ($USD)')
plt.title('Total Fare by City Type')
plt.savefig('analysis/Pyber_Fare_summary.png')


```
